```python
from google.colab import drive
drive.mount('/content/drive')

import scanpy as sc

adata = sc.read_h5ad('/content/drive/MyDrive/Tubitak-2209a/02_scRNA_analysis/scrna_filtered.h5ad')
print('Full:', adata.shape)

epi_types = ['Cancer Epithelial', 'Normal Epithelial']
epi = adata[adata.obs['celltype_major'].isin(epi_types)].copy()
print('Epithelial only:', epi.shape)
print(epi.obs['celltype_major'].value_counts())
```

```python
import numpy as np, pandas as pd

print('Max value (log check, expect >50 for raw counts):', epi.X.max())

expr = pd.DataFrame(epi.X.toarray().T, index=epi.var_names, columns=epi.obs_names)
print('Expression table (genes x cells):', expr.shape)

input_path = '/content/cytotrace2_input_expression.txt'
expr.to_csv(input_path, sep='\t')
print('Saved:', input_path)
```

```python
import pandas as pd

anno = pd.DataFrame({'cell_id': epi.obs_names, 'celltype': epi.obs['celltype_major'].values})
anno_path = '/content/cytotrace2_annotation.txt'
anno.to_csv(anno_path, sep='\t', index=False)
print('Annotation saved:', anno.shape)

from cytotrace2_py.cytotrace2_py import cytotrace2

results = cytotrace2(
    '/content/cytotrace2_input_expression.txt',
    annotation_path=anno_path,
    species='human',
    seed=14
)
print('DONE')
print(results.shape)
results.head()
```

```python
import os

results['celltype_major'] = epi.obs['celltype_major'].reindex(results.index).values
assert results.shape[0] == 28844, results.shape[0]
assert results['CytoTRACE2_Score'].notna().all(), "NaN in scores!"

out_dir = '/content/drive/MyDrive/Tubitak-2209a/02_scRNA_analysis'
out_path = os.path.join(out_dir, 'cytotrace2_epithelial_results.csv')
results.to_csv(out_path)
print('Saved:', out_path)

reloaded = pd.read_csv(out_path, index_col=0)
print('Reloaded:', reloaded.shape)
print(reloaded['CytoTRACE2_Potency'].value_counts())
print('Score range:', round(reloaded['CytoTRACE2_Score'].min(),3), '-', round(reloaded['CytoTRACE2_Score'].max(),3))
```
