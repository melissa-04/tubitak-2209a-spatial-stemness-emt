```python
import glob

pattern = '/content/metadata_extracted/metadata/*_metadata.csv'
files = sorted(glob.glob(pattern))
print(f'{len(files)} dosya bulundu:')
for f in files:
    print(' ', os.path.basename(f))

dfs = []
for f in files:
    temp = pd.read_csv(f)
    dfs.append(temp)

meta_all = pd.concat(dfs, ignore_index=True)
print('\nBirleşik tablo boyutu:', meta_all.shape)

print('\nTüm veri setindeki Classification değerleri:\n')
print(meta_all['Classification'].value_counts())
```

```python
import pandas as pd

def classify_region(label):

    if pd.isna(label):
        return 'missing'

    if label == 'DCIS':
        return 'in_situ'
    if label in ['Necrosis', 'Artefact', 'Uncertain', 'TLS',
                 'Cancer trapped in lymphocyte aggregation']:
        return 'exclude'

    if 'Invasive cancer' in label:
        return 'tumor'

    return 'non_tumor'

meta_all['region_class'] = meta_all['Classification'].apply(classify_region)


print('Standart grup dağılımı:\n')
print(meta_all['region_class'].value_counts())
print('\nToplam:', len(meta_all))


print('\nEtiketi boş (missing) spot:', (meta_all['region_class'] == 'missing').sum())


print('\nHam etiket → grup eşleşmesi:\n')
print(meta_all.groupby('region_class')['Classification'].unique())
```

```python
n_nan = meta_all['Classification'].isna().sum()
print('Classification NaN sayısı:', n_nan)

print('\nNaN spotların hasta dağılımı:')
print(meta_all[meta_all['Classification'].isna()]['patientid'].value_counts())
```

```python

meta_all['exclude_reason'] = ''
meta_all.loc[meta_all['region_class'] == 'missing', 'exclude_reason'] = 'no_pathology_label'
meta_all.loc[meta_all['region_class'] == 'exclude', 'exclude_reason'] = meta_all['Classification']

meta_all['region_class'] = meta_all['region_class'].replace('missing', 'exclude')

print('Güncel grup dağılımı:\n')
print(meta_all['region_class'].value_counts())
print('\nToplam:', len(meta_all))

print('\nDışlama gerekçeleri (exclude içinde):\n')
print(meta_all[meta_all['region_class'] == 'exclude']['exclude_reason'].value_counts())
```

```python
spatial_archive = os.path.join(RAW, 'spatial.tar.gz')

with tarfile.open(spatial_archive, 'r:gz') as tar:
    members = tar.getnames()

print(f'Arşivde {len(members)} öğe var.\n')
for m in members:
    print(m)
```

```python
spatial_dir = '/content/spatial_extracted'
with tarfile.open(spatial_archive, 'r:gz') as tar:
    tar.extractall(path=spatial_dir, filter='data')
print('Açıldı:', spatial_dir)

col_names = ['barcode', 'in_tissue', 'array_row', 'array_col', 'pxl_row', 'pxl_col']
test_path = '/content/spatial_extracted/spatial/1142243F_spatial/tissue_positions_list.csv'

pos = pd.read_csv(test_path, header=None, names=col_names)

print('\nBoyut:', pos.shape)
print('\nİlk 5 satır:')
print(pos.head())

print('\nin_tissue dağılımı:')
print(pos['in_tissue'].value_counts())
```

```python
col_names = ['barcode', 'in_tissue', 'array_row', 'array_col', 'pxl_row', 'pxl_col']


pos_files = sorted(glob.glob('/content/spatial_extracted/spatial/*_spatial/tissue_positions_list.csv'))
print(f'{len(pos_files)} koordinat dosyası bulundu.\n')

pos_dfs = []
for path in pos_files:

    folder = os.path.basename(os.path.dirname(path))
    patient = folder.replace('_spatial', '')

    p = pd.read_csv(path, header=None, names=col_names)
    p = p[p['in_tissue'] == 1].copy()
    p['patientid'] = patient

    print(f'{patient:12s}: {p.shape[0]} doku spotu')
    pos_dfs.append(p)


pos_all = pd.concat(pos_dfs, ignore_index=True)
print('\nBirleşik koordinat tablosu:', pos_all.shape)
print('\nİlk 3 satır:')
print(pos_all.head(3))
```

```python

meta_all = meta_all.rename(columns={'Unnamed: 0': 'barcode'})
print('Metadata kolonları:', meta_all.columns.tolist())

pos_subset = pos_all[['barcode', 'patientid', 'array_row', 'array_col', 'pxl_row', 'pxl_col']]

combined = meta_all.merge(
    pos_subset,
    on=['barcode', 'patientid'],
    how='left'
)


print('\nBirleşik tablo boyutu:', combined.shape)
print('(satır sayısı 15611 KALMALI — metadata patron)')

missing_coords = combined['array_row'].isna().sum()
print('\nKoordinatı eşleşmeyen spot:', missing_coords)
print('(0 olmalı — her metadata spotunun koordinatı bulunmalı)')
```

```python

unmatched = combined[combined['array_row'].isna()]

print('Eşleşmeyen spot sayısı:', len(unmatched))
print('\nHangi hastalarda?')
print(unmatched['patientid'].value_counts())

print('\nBu 10 spot (barcode + hasta + etiket):')
print(unmatched[['barcode', 'patientid', 'region_class']].to_string())


print('\nEşleşmeyen barcode uzunlukları:')
print(unmatched['barcode'].str.len().value_counts())


print('\nKoordinat tablosundaki barcode uzunlukları:')
print(pos_all['barcode'].str.len().value_counts())
```

```python

no_coord_mask = combined['array_row'].isna()
print('Koordinatsız spot:', no_coord_mask.sum())

combined.loc[no_coord_mask, 'exclude_reason'] = (
    combined.loc[no_coord_mask, 'exclude_reason'].astype(str) + ' + coordinate_missing'
)

combined['analyze'] = (combined['region_class'] != 'exclude') & (combined['array_row'].notna())


print('\nanalyze bayrağı dağılımı:')
print(combined['analyze'].value_counts())

print('\nAnalize girecek spotların region_class dağılımı:')
print(combined[combined['analyze']]['region_class'].value_counts())

print('\nKoordinatsız 10 spotun şu anki gerekçeleri:')
print(combined.loc[no_coord_mask, ['patientid', 'region_class', 'exclude_reason']].to_string())
```

```python
# ---- BÖLÜM 1: Özet bakış (kontrol) ----
print('='*60)
print('AŞAMA 0 ÖZET')
print('='*60)

print('\n1) Toplam spot:', len(combined))
print('   Analize girecek (analyze=True):', combined['analyze'].sum())
print('   Dışlanan (analyze=False):', (~combined['analyze']).sum())

print('\n2) Hasta başına ANALİZ spotu (tumor/non_tumor/in_situ):')
analyze_df = combined[combined['analyze']]
print(analyze_df.groupby('patientid').size())

print('\n3) Hasta başına TÜMÖR spotu (asıl ilgilendiğimiz):')
tumor_df = combined[combined['region_class'] == 'tumor']
print(tumor_df.groupby('patientid').size())

print('\n4) Hasta + alt tip eşleşmesi:')
print(combined.groupby('patientid')['subtype'].first())

# ---- BÖLÜM 2: Drive'a kaydet ----
import os

output_dir = '/content/drive/MyDrive/Tubitak-2209a/01_processed_data'
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, 'combined_metadata_spatial.csv')
combined.to_csv(output_path, index=False)

print('\n' + '='*60)
print('KAYDEDİLDİ:', output_path)
print('Boyut:', combined.shape)
print('Kolonlar:', combined.columns.tolist())
```
